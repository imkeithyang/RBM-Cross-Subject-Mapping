```python
import pickle
from scipy.io import loadmat
import numpy as np
#hiwa = pickle.load(open("Results/08302023/hiwa_transfer_moth_1", "rb"))
FD = loadmat("Results_CV/NoHiWA_SCENARIO_I_F_OneMissingSubj_sigma0.0025_PCAThreshold10_noHidUnit15_lr0.005_tol0.1_TrainTestRatio0.8")
CD = loadmat("Results_CV/NoHiWA_SCENARIO_I_CD_OneMissingSubj_sigma0.0025_PCAThreshold10_noHidUnit15_lr0.005_tol0.1_TrainTestRatio0.8")
FDk5 = loadmat("Results_CV/k=5NoHiWA_SCENARIO_I_F_OneMissingSubj_sigma0.0025_PCAThreshold10_noHidUnit15_lr0.005_tol0.1_TrainTestRatio0.8")
CDk5 = loadmat("Results_CV/k=5NoHiWA_SCENARIO_I_CD_OneMissingSubj_sigma0.0025_PCAThreshold10_noHidUnit15_lr0.005_tol0.1_TrainTestRatio0.8")

#procrustes = loadmat("Results/08302023/procrustes_SCENARIO_I_CD_OneMissingSubj_sigma0.0025_PCAThreshold10_noHidUnit15_lr0.005_tol0.1_TrainTestRatio0.5")
```

```python
all_result = []
all_label = []
moth_label = []
Z = [1,2,3,4,6,7,8,9,10]
C = [1,2,3,4,5,6,7,8,9]
seed_cutoff = 5
import pandas as pd
for index,i in enumerate(Z):
    #hiwa = pickle.load(open("Results/08302023/hiwa_transfer_moth_{}".format(i), "rb"))
    LDA = list(FD["Rate_LDA"]["moth_{}".format(i)][0][0][-1,0:seed_cutoff])
    Fisher = list(FD["Rate_FRBM"]["moth_{}".format(i)][0][0][-1,0:seed_cutoff])
    Fisherk5 = list(FDk5["Rate_FRBM"]["moth_{}".format(i)][0][0][-1,0:seed_cutoff])
    MLCD = list(CD["Rate_MLRBM"]["moth_{}".format(i)][0][0][-1,0:seed_cutoff])
    MLCDk5 = list(CDk5["Rate_MLRBM"]["moth_{}".format(i)][0][0][-1,0:seed_cutoff])
    No_T = list(CD["Rate_NO"]["moth_{}".format(i)][0][0][0,0:seed_cutoff])
    #proco = list(procrustes["Rate_procrustes"]["moth_{}".format(i)][0][0][0,0:seed_cutoff])
    #hiwa_result = []
    #for j, seed_result in enumerate(hiwa):
    #    hiwa_result.append(np.mean(seed_result[0]))
    all_result += LDA + Fisher +Fisherk5 + MLCD +MLCDk5 + No_T #hiwa_result + proco 
    
    all_label += ["Subject Specific"]*seed_cutoff + ["RBM-FD k=1"]*seed_cutoff +\
            ["RBM-FD k=5"]*seed_cutoff + ["RBM-CD"]*seed_cutoff + ["RBM-CD k=5"]*seed_cutoff + ["No Transfer"]*seed_cutoff#+ ["HiWA"]*seed_cutoff + \
            #["Procrustes"]*seed_cutoff 
            
    moth = ["Moth {}".format(C[index])]
    moth_label += moth*(seed_cutoff*6) 
df = pd.DataFrame({"Decoding Accuracy":all_result, "Method": all_label, "": moth_label})
    
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 14

plt.rc('font', size=BIGGER_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=16)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=BIGGER_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

plt.figure(figsize=(17,5), dpi=200)
ax = sns.boxplot(data=df, x="", y="Decoding Accuracy", hue="Method", linewidth=0.8,width=0.8,
              showfliers = False)
ax.legend(loc='lower center', bbox_to_anchor=(0.5, -0.25),ncol=6)
plt.savefig("Scenario1_CVk=5.eps", bbox_inches="tight")
```
